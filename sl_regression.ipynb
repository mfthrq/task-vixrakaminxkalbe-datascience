```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Dataset CCV 5 - Supervised Learning Regression.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
plt.figure(figsize=(12,5))
sns.pairplot(data, x_vars=['Diabetasol Coklat 1 kg'], y_vars=['Daily Sales'], height=5, kind='scatter')
plt.xlabel('Diabetasol Coklat 1 kg')
plt.ylabel('Daily Sales')
plt.title('Scatterplot Penjualan Diabetasol Coklat 1 kg dengan Total Penjualan Harian')
```

```python
data.dropna(subset = 'Diabetasol Coklat 1 kg', inplace=True)
```

```python
X = data['Diabetasol Coklat 1 kg']
X.head()
```

```python
y = data['Daily Sales']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
X_train = X_train.to_numpy()
X_test = X_test.to_numpy()
```

```python
X_train = X_train.reshape(X_train.shape[0],1)
X_test = X_test.reshape(X_test.shape[0],1)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
c = [i for i in range (1, len(y_test)+1, 1)]
plt.plot(c, y_test, color='r', linestyle='-', label='Actual Data')
plt.plot(c, y_pred, color='b', linestyle='-', label='Predicted Data')
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
plt.legend()
plt.title("Perbandingan antara data aktual dan data prediksi")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error", mse)
```

```python
print("Intercept of the model : ", lr.intercept_)
print("Coefficient of the model : ", lr.coef_)
```
